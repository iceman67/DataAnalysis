```python
import pandas as pd
import requests
import json
import xmltodict
import datetime
```

```python
url='http://openapi.seoul.go.kr:8088/456b6a414c73736137367a56446671/xml/LOCALDATA_010101/1/1000/'
```

```python
requests.get(url)
```

```python
data=requests.get(url).text
```

```python
json.dumps(xmltodict.parse(data))
```

```python
result=json.loads(json.dumps(xmltodict.parse(data)))
```

```python
df=pd.DataFrame(result['LOCALDATA_010101']['row'])
```

```python
df=df[['APVPERMYMD', 'TRDSTATENM', 'DCBYMD', 'SITEWHLADDR', 'RDNWHLADDR', 'BPLCNM', 'UPTAENM']]
df.columns=['', '', '', '', '', '', '']
```

```python
df[''] = df[''].str.split(" ").str.get(1)
df['']=df[''].fillna(value=df[""].str.split(" ").str.get(1))
df.loc[df[''].str.contains(''), ''] = ''
df['']=pd.to_datetime(df[''])
df['']=pd.to_datetime(df[''])
df['']=df[''].dt.year
df['']=df[''].dt.year
```

```python
df.to_csv('.csv',encoding='euc-kr', index=False)
```
