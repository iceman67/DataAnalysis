```python
import pandas as pd
import requests
import json
import xmltodict
import datetime
```

```python
url='http://openapi.seoul.go.kr:8088/456b6a414c73736137367a56446671/xml/LOCALDATA_010101/1/1000/'
```

```python
requests.get(url)
```

```python
data=requests.get(url).text
```

```python
json.dumps(xmltodict.parse(data))
```

```python
result=json.loads(json.dumps(xmltodict.parse(data)))
```

```python
df=pd.DataFrame(result['LOCALDATA_010101']['row'])
```

```python
df=df[['APVPERMYMD', 'TRDSTATENM', 'DCBYMD', 'SITEWHLADDR', 'RDNWHLADDR', 'BPLCNM', 'UPTAENM']]
df.columns=['인허가일자', '영업상태명', '폐업일자', '지번주소', '도로명주소', '사업자명', '업태구분명']
```

```python
df['행정구'] = df['도로명주소'].str.split(" ").str.get(1)
df['행정구']=df['행정구'].fillna(value=df["지번주소"].str.split(" ").str.get(1))
df.loc[df['업태구분명'].str.contains('요양병원'), '업태구분명'] = '요양병원'
df['인허가일자']=pd.to_datetime(df['인허가일자'])
df['폐업일자']=pd.to_datetime(df['폐업일자'])
df['인허가일자']=df['인허가일자'].dt.year
df['폐업일자']=df['폐업일자'].dt.year
```

```python
df.to_csv('병원.csv',encoding='euc-kr', index=False)
```
