```
import os
import requests
import pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```
def sido_data_search(yyyymm):
  date = yyyymm

  year = []
  mon = []
  sido_name = []
  sido_expCnt = []
  sido_expUsdAmt = []
  sido_impCnt = []
  sido_impUsdAmt = []
  sido_balPayments = []
  
  sido_url = 'http://openapi.customs.go.kr/openapi/service/newTradestatistics/getsidotradeList?searchBgnDe=' + yyyymm + '&searchEndDe=' + yyyymm + '&serviceKey='
  sido_key = '1P0EFkE9zCTMG4PND3EGAQN%2FKVg%2BrbQguzgsLoWh%2BIzapXwOqSfvJKivIl7oAK6gwia26UZhtGZ%2Fg4r1gJnWgQ%3D%3D'

  sido_request_api = sido_url + sido_key
  sido_res = requests.get(sido_request_api)
  sido_soup = BeautifulSoup(sido_res.content,'xml')
  sido_item = sido_soup.find_all('item')

  for i in sido_item:
    if i.sidoNm:
      if len(i.sidoNm.get_text()) > 4:
        sido_name.append(i.sidoNm.get_text())
        sido_expCnt.append(i.expCnt.get_text().strip().replace(',',''))
        sido_impCnt.append(i.impCnt.get_text().strip().replace(',',''))
        sido_expUsdAmt.append(i.expUsdAmt.get_text().strip().replace(',',''))
        sido_impUsdAmt.append(i.impUsdAmt.get_text().strip().replace(',',''))
        year.append(date[0:4])
        mon.append(date[4:6])

  sum_data = {
      '년': year, 
      '월': mon,
      '시도명': sido_name,
      '수출건수(건)': sido_expCnt, 
      '수입건수(건)': sido_impCnt, 
      '수출금액(USE 1000)': sido_expUsdAmt, 
      '수입금액(USE 1000)': sido_impUsdAmt, 
  }

  return sum_data
```

```
def sido_data_price(yyyymm):
  date = yyyymm
  sum_data = {}
  if os.path.exists('sido_impexp_data.csv'):
    os.remove('sido_impexp_data.csv')

  for ym in date:
    base = datetime.strptime(ym, '%Y%m')
    date_list = [(base + relativedelta(months=x)).strftime('%Y%m') for x in range(0, 6)] 

    for i in date_list:
      sido_data = sido_data_search(i)

      df = pd.DataFrame(sido_data)
      
      if not os.path.exists('sido_impexp_data.csv'):
        df.to_csv('sido_impexp_data.csv', index = False, mode = 'w', encoding = 'utf-8-sig')
      else:
        df.to_csv('sido_impexp_data.csv', index = False, mode = 'a', encoding = 'utf-8-sig', header = False)  
```

```
ymd_sido = ['201901', '202001']
sido_data = sido_data_price(ymd_sido)
```

```
sido_data_elections = pd.read_csv(sido_data_url)

sido_2019 = sido_data_elections[sido_data_elections['년'] == 2019]
sido_2020 = sido_data_elections[sido_data_elections['년'] == 2020]


sido_local_2019 = []
sido_local_2020 = []
sido_column = []
exp_sum_2019 = []
exp_sum_2020 = []

for idx, i in enumerate(sido_2020):
  sido_column.append(i) if idx > 2 else 0

print(sido_column)

for i in range(2):
    for column in sido_column:
      exp_sum_2019.append(sido_2019[column].sum()) if i == 0 else exp_sum_2020.append(sido_2020[column].sum())
```

```
ax = plt.subplot(121)
ax2 = plt.subplot(122)

ax.set_title('19/20년 상반기 수출입 동향')
ax2.set_title('19/20년 상반기 수출입 동향')

plt.rc('font', family='NanumBarunGothic')
plt.rcParams["figure.figsize"] = (20,8) 
plt.rcParams.update({'font.size': 18})

def create_x(t, w, n, d):
    return [t*x + w*n for x in range(d)]

ax.set_xticklabels(['수출', '수입'])
ax2.set_xticklabels(['수출', '수입'])

value_a_x = create_x(2, 0.8, 1, len(sido_column[0:2]))
value_b_x = create_x(2, 0.8, 2, len(sido_column[0:2]))
value_a_x2 = create_x(2, 0.8, 1, len(sido_column[2:4]))
value_b_x2 = create_x(2, 0.8, 2, len(sido_column[2:4]))

middle_x1 = [(a+b)/2 for (a,b) in zip(value_a_x, value_b_x)]
middle_x2 = [(a+b)/2 for (a,b) in zip(value_a_x2, value_b_x2)]

ax.bar(value_a_x, exp_sum_2019[0:2]) 
ax.bar(value_b_x, exp_sum_2020[0:2])
ax2.bar(value_a_x2, exp_sum_2019[2:4]) 
ax2.bar(value_b_x2, exp_sum_2020[2:4])


for i, p in enumerate(ax.patches):
    left, bottom, width, height = p.get_bbox().bounds
    ax.annotate(int(height), (left+width - 0.85/2, height * 1.007), ha='center', fontSize = 18, color = 'blue' if i<2 else 'red')
for i, p in enumerate(ax2.patches):
    left, bottom, width, height = p.get_bbox().bounds
    ax2.annotate(int(height), (left+width - 0.85/2, height * 1.007), ha='center', fontSize = 18, color = 'blue' if i<2 else 'red')

ax.set_xlabel('수출/수입')
ax.set_ylabel('건수(백 만)')
ax.legend([2019, 2020])
ax.set_xticks(middle_x1)
ax.grid(True, linestyle ='--')

ax2.set_xlabel('수출/수입')
ax2.set_ylabel('금액(USE 1000)')
ax2.legend([2019, 2020])
ax2.set_xticks(middle_x2)
ax2.grid(True, linestyle ='--')


plt.show()
```

```
sido_2019_exp = []
sido_2020_exp = []

for i in sido_2019['월'].unique():
  sido_2019_exp.append(sido_2019[sido_2019['월'] == int(i)]['수출금액(USE 1000)'].sum())
  sido_2020_exp.append(sido_2020[sido_2020['월'] == int(i)]['수출금액(USE 1000)'].sum())
```

```
plt.rc('font', family='NanumBarunGothic')  
plt.rcParams["figure.figsize"] = (20,8) 
plt.rcParams.update({'font.size': 20})

plt.title('19/20년 상반기 월별 수출금액')

mon = sido_2019['월'].unique()

plt.plot(mon, sido_2019_exp)
plt.scatter(mon, sido_2019_exp, c = 'b', s = 100)
plt.plot(mon, sido_2020_exp)
plt.scatter(mon, sido_2020_exp, c = 'r', s = 100)

for i, v in enumerate(sido_2019_exp):
  plt.text(mon[i], v * 1.005, v, fontSize = 15, color = 'blue', horizontalalignment = 'center', verticalalignment = 'bottom')
for i, v in enumerate(sido_2020_exp):
  plt.text(mon[i], v* 1.008, v, fontSize = 15, color = 'red', horizontalalignment = 'center', verticalalignment = 'bottom')

plt.xlabel('월')
plt.ylabel('금액 (USE 1000)')
plt.legend([2019, 2020])
plt.grid(True, linestyle ='--')

plt.show()
```
