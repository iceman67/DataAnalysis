```python
import pandas as pd
```

```python
df=pd.read_csv('도로형태별.csv', encoding='cp949')
df
```

```python
df1=df.groupby(['시도','도로형태'])['사고건수'].sum().unstack()
df1
```

```python
df3=df1[['고가도로위', '교량위', '지하차도(도로)내', '터널안']]
df3
```

```python
df4=df1[['고가도로위', '교량위', '교차로내', '지하차도(도로)내', '터널안']]
df4
```

```python
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc
import matplotlib
font_location='c:/Windows/fonts/malgun.ttf' 
font_name=font_manager.FontProperties(fname=font_location).get_name()
matplotlib.rc('font',family=font_name)
```

```python
plt.figure(figsize=(16,9))
df3.plot(kind='bar', figsize=(25,9), width=0.6)
plt.title('2019년 시군구 도로형태별 교통사고 (교차로 제외)', size=35)
plt.grid(alpha=0.5)
plt.xlabel('시도', size=18)
plt.ylabel('사고건수', size=20)
plt.xticks(rotation=0, size=20)
plt.yticks(rotation=0, size=20)
plt.legend(fontsize=27)
```

```python
plt.figure(figsize=(16,9))
df4.plot(kind='bar', figsize=(25,9), width=0.8)
plt.title('2019년 시군구 도로형태별 교통사고 (교차로 포함)', size=35)
plt.xlabel('시도', size=18)
plt.ylabel('사고건수', size=20)
plt.xticks(rotation=0, size=20)
plt.yticks(rotation=0, size=20)
plt.grid(alpha=0.5)
plt.legend(fontsize=27)
```

```python
df_n=pd.read_csv('기상상태별.csv', encoding='cp949')
df_n
```

```python
dfn=df_n.groupby(['시도','기상상태'])['사고건수'].sum().unstack()
dfn
```

```python
dfn1=dfn[['기타/불명', '눈', '비', '안개', '흐림']]
dfn1
```

```python
plt.figure(figsize=(16,9))
dfn.plot(kind='bar', figsize=(25,9), width=0.8)
plt.title('2019년 시군구 기상상태별 교통사고 (맑음 포함)', size=35)
plt.grid(alpha=0.5)
plt.xlabel('시도', size=18)
plt.ylabel('사고건수', size=20)
plt.xticks(rotation=0, size=20)
plt.yticks(rotation=0, size=20)
plt.legend(fontsize=28)
```
