```python
#localurl='April_sensor_data1.csv'
localurl= 'https://raw.githubusercontent.com/iceman67/DataAnalysis/master/April_sensor_data.csv'
try:
    dateparse = lambda d: pd.datetime.strptime(d, '%Y-%m-%d %H:%M')
    df = pd.read_csv(localurl, parse_dates=['REG_DATE'], date_parser=dateparse)
except  OSError as err:
    print("OS error: {0}".format(err))
except ValueError as err:
    print("Value error: {0}".format(err))
```
